```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 任意の整数
m = 6
```

```python
# 符号長
n = 2**m - 1
# 情報数
k = n - m
```

```python
np.set_printoptions(precision=0,threshold=100000,linewidth=np.inf)
# 検査行列
H = np.zeros((m, n))
# 右側を単位行列にする
for i in range(m):
    H[i][k + i] = 1
# 左側はランダム
H_set = set()
H_set.add(0)
for i in range(m):
    H_set.add(1 << i)
for i in range(k):
    while True:
        h = np.random.randint(low=0, high=2, size=(m, 1))
        sum_h = 0
        for j in range(m):
            sum_h += h[j, 0] << j
        if sum_h not in H_set:
            H[:, i] = h.reshape(-1)
            H_set.add(sum_h)
            break
print("---検査行列---")
for h in H:
    print(np.int64(h), sep="\t")
print(np.linalg.matrix_rank(H[:, :k]))
# 情報列
x = np.random.randint(low=0, high=2, size=k)
print("---情報列---")
print(np.int64(x), sep="\t")
# 生成行列
G = np.concatenate([np.identity(k), H[:, :k].T], axis=1)
print("---生成行列---")
for g in G:
    print(np.int64(g), sep="\t")
# 符号列
w = x @ G % 2
print("---符号列---")
print(np.int64(w), sep="\t")
```

```python
# チェック(正しく伝送された場合)
result = H @ w.T % 2
print("---誤りなし---")
print(np.int64(result), sep="\t")
# チェック(1ビット誤り)
print("---1ビット誤り---")
for i in range(n):
    wi_old = w[i]
    w[i] = (w[i] + 1) % 2
    result = H @ w.T % 2
    print(f"{i+1}ビット目が誤った場合: {np.int64(result)}", sep="\t")
    w[i]=wi_old
```
